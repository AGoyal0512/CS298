```python
df['IncidentDate'] = pd.to_datetime(df['IncidentDate'])
```

```python
df['IncidentDate'][0].day
```

```python
for i in range(df.shape[0]):
    df['Address'][i] = df['Address'][i].rstrip()
```

```python
for i in range(df.shape[0]):
    df['Address'][i] = df['Address'][i]+', Madison, Wisconsin'
```

```python
df['Address']
```

```python
df['Day'] = pd.Series(np.zeros(df.shape[0]))
df['Month'] = pd.Series(np.zeros(df.shape[0]))
df['Year'] = pd.Series(np.zeros(df.shape[0]))
df['Hour'] = pd.Series(np.zeros(df.shape[0]))
```

```python
for i in range(df.shape[0]):
    df['Day'][i] = df['IncidentDate'][i].day
    df['Month'][i] = df['IncidentDate'][i].month
    df['Year'][i] = df['IncidentDate'][i].year
    df['Hour'][i] = df['IncidentDate'][i].hour
    if df['IncidentDate'][i].minute >= 30:
        df['Hour'][i] = df['IncidentDate'][i].hour + 1
```

```python
df = df.drop('IncidentDate', axis = 1)
df = df.astype({'Day': 'int32', 'Month': 'int32', 'Year': 'int32', 'Hour': 'int32'})
df = df.drop('DateModified', axis = 1)
```
